```python
#@title Bài 1
num_list = [3 , 4 , 5 , 1 , -44 , 5 ,10 , 12 ,33 , 1]
def bai_1(k):
  max_list = []
  n=len(num_list)-k+1
  for i in range(n):
    k_list = []
    for j in range(k):
      k_list.append(num_list[i+j])
    print(k_list, end=' ')
    print(f' => Giá trị max là: {max(k_list)}')
    max_list.append(max(k_list))
  return max_list
bai_1(3)
```

```python
#@title Bài 2
def bai_2(string):
  dic = {}
  for i in string:
    if i not in dic:
      dic[i] = 1
    else:
      dic[i] += 1
  return dic
print(bai_2('smiles'))
```

```python
#@title Bài 3
txt_path = 'P1_data.txt'

def bai_3(txt_path):
  import re
  with open(txt_path, 'r') as file:
    content = file.read()
  # print(content)
  content = re.sub(r'\s+', ' ', content)
  content = content + ' '
  # print(content)
  key = ''
  dic = {}
  for i in content:
    if i != ' ':
      key += i
    else:
      if key not in dic:
        dic[key] = 1
      else:
        dic[key] += 1
      key = ''
  return dic
print(bai_3(txt_path)['man'])
```

```python
#@title Bài 4
source = 'hola'
target = 'hello'

def bai_4(source,target):
  key = [source,target]
  def sub_cost(a,b):
    if a==b:
      return 0
    else:
      return 1
  dic = []
  row = len(source)+1
  col = len(target)+1
  for i in range(row):
    dic.append([])
    for j in range(col):
      dic[i].append(0)
  for i in range(1,row):
    dic[i][0] = i
  for j in range(1,col):
    dic[0][j] = j
  for i in range(1,row):
    for j in range(1,col):
      dic[i][j] = min(dic[i][j-1]+1,dic[i-1][j]+1,dic[i-1][j-1]+sub_cost(key[0][i-1],key[1][j-1]))
      print(dic[i][j], end=" ")
    print()
  print(f'Khoảng cách chỉnh sửa tối thiểu Levenshtein : {dic[i][j]}')
bai_4(source,target)
```
